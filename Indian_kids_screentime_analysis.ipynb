#Exploratory Data Analysis

1.   Gender



```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x='Gender', data=df, palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs Avg_Daily_Screen_Time_hr

sales_gen = df.groupby(['Gender'], as_index=False)['Avg_Daily_Screen_Time_hr'].sum().sort_values(by='Avg_Daily_Screen_Time_hr', ascending=False)
sns.barplot(x = 'Gender',y= 'Avg_Daily_Screen_Time_hr' ,data = sales_gen, palette='viridis')
```

2.   Age


```python
ax = sns.countplot(data = df, x = 'Age', hue = 'Gender', palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age'], as_index=False)['Avg_Daily_Screen_Time_hr'].sum().sort_values(by='Avg_Daily_Screen_Time_hr', ascending=False)
sns.barplot(x = 'Age',y= 'Avg_Daily_Screen_Time_hr' ,data=sales_age, palette='viridis')
```



3. 	Health_Impacts



```python
sns.set(rc={'figure.figsize':(10,6)})
ax = sns.countplot(data=df, x='Health_Impacts', palette='viridis')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
for container in ax.containers:
    ax.bar_label(container, padding=3)

plt.tight_layout()
plt.show()
```

4. Educational_to_Recreational_Ratio

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Educational_to_Recreational_Ratio', palette='viridis')

for bars in ax.containers:
    ax.bar_label(bars)
```

5. Urban_or_Rural

```python
sales_state = df.groupby(['Urban_or_Rural'], as_index=False)['Age'].sum().sort_values(by='Age', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Urban_or_Rural',y= 'Age', palette='viridis')
```

# Conclusion:

Based on recent comprehensive research on Indian children's screen time, it is evident that children under five are spending an average of about 2.2 hours daily on screens, which is twice the safe limit recommended by global and national health authorities like the WHO and Indian Academy of Pediatrics. Alarmingly, even infants under two years of age show average daily screen exposure of around 1.2 hours, despite guidelines advising zero screen time for this age group. This excessive screen exposure is linked with several developmental and health risks, including delayed language development, reduced cognitive skills, impaired social interaction, poor sleep quality, increased risk of obesity, and concentration difficulties. The broad prevalence of such high screen usage among young Indian children highlights a critical public health concern, emphasizing the urgent need for parents, caregivers, and policymakers to create structured screen time limits, designate tech-free zones, and encourage more offline play and family engagement. Without these measures, the long-term physical, emotional, and cognitive well-being of Indian children could be seriously compromised, making awareness and intervention efforts crucial at this stage.
